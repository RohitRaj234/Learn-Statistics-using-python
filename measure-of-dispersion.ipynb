**Method 1: Variance implemented using manually**

```python
data = [70, 75, 80, 85, 90]  
mean =sum(data)/len(data)
squared_diffs = [(x - mean) ** 2 for x in data]
pop_variance=sum(squared_diffs)/len(data)
sample_variance = sum(squared_diffs) / (len(data) - 1)
print(mean)
print(pop_variance)
print(sample_variance)
```

**Method 2: Variance using numpy**

```python
data = [70, 75, 80, 85, 90]  
sam_var=np.var(data,ddof=1)
pop_var=np.var(data,ddof=0)
print(sam_var)
print(pop_var)
```

**Method 3: variance using pandas**

```python
data = [70, 75, 80, 85, 90] 
df = pd.Series(data)
print(df.var(ddof=1))          
print(df.var(ddof=0)) 
```

**Method 4: variance using statistics**

```python
import statistics as stats

data = [70, 75, 80, 85, 90]

sam_var = stats.variance(data)     
pop_var = stats.pvariance(data)    

print("Sample Variance:", sam_var)
print("Population Variance:", pop_var)
```

**Method 1 Standard Deviation Manual Implementation in Python**

```python
import math

data = [70, 75, 80, 85, 90]
mean = sum(data) / len(data)


squ_diff = [(x - mean) ** 2 for x in data]


pop_var = sum(squ_diff) / len(data)
pop_std_dev = math.sqrt(pop_var)

sample_variance = sum(squ_diff) / (len(data) - 1)
sample_std_dev = math.sqrt(sample_variance)

print(sample_std_dev)
print(pop_std_dev)
```

**Method 2 : Standard Devitation using Numpy**

```python
import numpy as np

data = [70, 75, 80, 85, 90]

print("Population Std Dev:", np.std(data))             
print("Sample Std Dev    :", np.std(data, ddof=1))
```

**Method 3: Standard Devitation using pandas**

```python
import pandas as pd
data = [70, 75, 80, 85, 90]
df=pd.Series(data)
sample_std=df.std()
pop_std = df.std(ddof=0)
print(sample_std)
print(pop_std)
```

**Method 4 : Standard Devitation using statistics**

```python
import statistics as stats
data=[70,75,80,85,90]
sample_std=stats.stdev(data)
pop_std=stats.pstdev(data)
print(sample_std)
print(pop_std)
```

**Mean Absolute Deviatation**
✅ Simpler and more robust to outliers than standard deviation
✅ Good for understanding average deviation without squaring differences
✅ Useful when you care about the actual difference, not squared or complex forms



```python
data = [70, 75, 80, 85, 90]
mean=sum(data)/len(data)
mad=sum(abs(x-mean) for x in data)/len(data)
print(mad)
```

```python
import numpy as np 
data = [70, 75, 80, 85, 90]
mad=np.mean(np.abs(data-np.mean(data)))
print(mad)
```


**IQR (Interquartile Range) is a measure of dispersion that shows the spread of the middle 50% of a dataset.
IQR=Q3−Q1Q1 (1st Quartile): 25% of the data falls below this value.Q3 (3rd Quartile): 75% of the data falls below this value.**


```python
import numpy as np

data = [70, 75, 80, 85, 90]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```
